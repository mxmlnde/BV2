```python
# Erzeugung Testdaten
rows = 500
test_data = np.zeros((rows, 3))

for i in range(rows):
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    sum = num1 + num2
    test_data[i] = [num1, num2, sum]

data_train, data_test = train_test_split(test_data, test_size=0.2, random_state=42)
```

```python
w = np.array([0.0001, 0.0001]) # Gewichte
n = 0.01 # Lernrate
b = 1 # Brauch man ein Bias?
storeW = [np.copy(w)]

# Einfaches Netz
def einfaches_Netz(data, n, durchlauf):
  for i in range(durchlauf):
    for i in range(data.shape[0]): # data.shape[0]
      storeW.append(w)
      # y gerechnet/gew端nscht
      y_d = np.dot(data[i][:2], w)
      #print(y_d)
      y = data[i][2]
      #print(y)

      # Fehler
      l = 0.5 * (y_d - y) ** 2
      # print(l)

      # Gradienten Abstieg
      gradient_descent = [(y_d - y) * data[i][0], (y_d - y) * data[i][1]]

      # Gewichte werden angepasst
      w[0] = w[0] - n * (gradient_descent[0]/1000)
      w[1] = w[1] - n * (gradient_descent[1]/1000)
      storeW.append(np.copy(w))
```
